* Adding appropriate titles

Please create code cells below this to create your visualizations.

```python
# prompt: Please create a chart between type and length.

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x='type', y='length', data=df)
plt.title('Average Length by Type of Dinosaur')
plt.xlabel('Type')
plt.ylabel('Average Length (meters)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# prompt: Please create a new column called "period name" by extracting the first two words of the period column.

df['period name'] = df['period'].str.split().str[:2].str.join(' ')
```

```python
df.info()
```

```python
df["period name"]
```

```python
# prompt: Please make a graph between length and period name/

plt.figure(figsize=(12, 6))
sns.boxplot(x='period name', y='length', data=df)
plt.title('Distribution of Lengths by Period')
plt.xlabel('Period')
plt.ylabel('Length (meters)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

The bar plot visualizes the average length of dinosaurs categorized by their 'type'. The x-axis represents the different dinosaur types, and the y-axis shows the average length in meters for each type. This plot helps to compare the average sizes across different classifications of dinosaurs.
